# **Langton's Ant Simulation**

**Introduction**

This Jupyter Notebook simulates Langton's Ant, a simple cellular automaton that exhibits complex emergent behavior. Langton's Ant moves on a grid of black and white squares, following a set of rules:

* **Rule 1:** If the ant is on a white square, it turns right, flips the square to black, and moves forward.
* **Rule 2:** If the ant is on a black square, it turns left, flips the square to white, and moves forward.

Despite the simplicity of these rules, Langton's Ant can produce intricate patterns and behaviors, demonstrating the concept of emergent complexity.

This simulation provides a visual representation of the ant's movements and the resulting grid patterns.


## Step 1: Import Necessary Libraries
1. matplotlib.pyplot: Used for creating plots and visualizations.
2. numpy: Provides efficient numerical operations for array manipulation.
3. random: Used to generate random numbers for the initial grid.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Step 2: Define Constants
width and height: Set the dimensions of the grid.

```python
width, height = 100, 100  # Grid size
```

## Step 3: Initialize Grid and Ant
1. Create grid: Generate a random grid of black and white squares using np.random.randint.
2. Set initial ant position: Place the ant at the center of the grid.
3. Initialize direction: Set the ant's initial direction (0: up, 1: right, 2: down, 3: left).

```python
grid = np.random.randint(0, 2, (width, height))  # Create a random grid
ant_x, ant_y = width // 2, height // 2  # Set initial ant position
direction = 0  # 0: up, 1: right, 2: down, 3: left
```

## Step 4: Simulation Loop
Iterate over the simulation: Run the simulation for a specified number of iterations.
Update ant direction and grid:
If the ant is on a white square, turn right, flip the square to black, and move forward.
If the ant is on a black square, turn left, flip the square to white, and move forward.
Update ant position: Move the ant in the current direction.
Handle boundary conditions: Implement wrapping around for the grid boundaries.
Track ant path: Append the current ant position to the path list.

```python
path = [(ant_x, ant_y)]  # Track ant's path
for _ in range(10000):
    # Update ant direction and grid
    if grid[ant_y][ant_x] == 0:
        direction = (direction + 1) % 4
        grid[ant_y][ant_x] = 1
    else:
        direction = (direction - 1) % 4
        grid[ant_y][ant_x] = 0

    # Update ant position
    if direction == 0:
        ant_y -= 1
    elif direction == 1:
        ant_x += 1
    elif direction == 2:
        ant_y += 1
    else:
        ant_x -= 1

    # Handle boundary conditions (wrap around)
    if ant_y < 0:
        ant_y = height - 1
    elif ant_y >= height:
        ant_y = 0
    if ant_x < 0:
        ant_x = width - 1
    elif ant_x >= width:
        ant_x = 0

    path.append((ant_x, ant_y))
```

## Step 5: Visualize Results
1. Create subplots: Create two subplots for the grid and the ant's path.
2. Plot the grid: Display the final state of the grid using imshow.
3. Plot the ant's path: Plot the ant's trajectory using plot.
4. Show the plots: Display the visualizations.

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the grid
axs[0].imshow(grid, cmap='gray')
axs[0].set_title('Grid')

# Plot the ant's path
axs[1].plot(*zip(*path), marker='o', linestyle='-', color='red')
axs[1].set_title('Ant Path')

plt.show()
```

In summary, the code above creates two subplots:

* Grid Plot: Shows the black and white grid used in the simulation.
* Ant Path Plot: Displays the trajectory of the ant as it moves through the grid.
  
These plots provide a visual representation of the Langton's Ant simulation, allowing you to analyze the emergent behavior and patterns that arise from the ant's interactions with the grid.
